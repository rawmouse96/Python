- indexing

```python
# 리스트에는 다양한 자료형의 데이터를 사용할 수 있음
name = ('bj', 'jj', 'jb', 'kh')
print(type(name), name[0])
person = ['James', 27, 175.3, '남자']
print(type(person), person[0])
```

```python
# 슬라이싱 가능
a = [3, 7, 13, 14, 19]
a[1:-1], a[:2], a[3:]
```
