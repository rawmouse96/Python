- 논리 연산자

```python
# 논리 곱(and)
t, f = True, False
f and f, f and t, t and f, t and t
```

```python
# 논리 합(or)
f or f, f or t, t or f, t or t
```

```python
# 논리 부정(not)
not f, not t
```

- 논리연산자와 비교연산자의 혼용

```python
import random
rd = random.random()                # 0 <= x < 1,실수 랜덤 기본형
rdint = random.randint(1, 6)        # 1 <= x <= 6,정수 x이상 y이하 범위지정
print(rd, rdint)
```

```python
# 주사위의 눈이 2이하 or 5이상
dice = random.randint(1, 6)
print(dice, dice <= 2 or dice >= 5)
```

```python
# 주사위의 눈이 2이상 and 5이하
print(dice, dice>=2 and dice<=5)
print(dice, 2<=dice<=5)
```

```python
# 주사위를 세번 던져서 모두 같은 수가 나왔는지 체크
dice1, dice2, dice3 = random.randint(1, 6), random.randint(1, 6), random.randint(1, 6)
print(dice1, dice2, dice3, dice1==dice2 and dice2==dice3)
print(dice1, dice2, dice3, dice1==dice2==dice3)
```

- Exclusive OR(XOR)
    * f xor f, t xor t --> False
    * f xor t, t xor f --> True

```python
import operator
operator.xor(False, True)

a, b = True, False
(a and not b) or (not a and b) 
```

- Bitwise logical operator
    - &     and
    - |     or
    - ~     not
    - ^     xor

```python
a, b = 0, 1
print(a & b, a | b, a ^ b)
```
