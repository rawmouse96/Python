## 5-3. 람다함수

- 익명 함수

```python
def plus_ten(x):
    return x + 10
```

```python
plus_ten(4)
```

```python
plus_ten = lambda x: x +10
plus_ten(4)
```

- map 함수에 적용

```python
list(map(lambda x: x+10, [1, 3, 5]))
```

```python
def power(x):
    return x*x
a = [1, 2, 3, 4, 5]
list(map(power(a), a))
list(map(lambda x: x*x, a))
```

- 람다 함수 만드는 법

```python
f = lambda x: print(x)
f('Hello')
```

```python
f = lambda x: x ** 3
f(10)
```

```python
# 홀수면 세제곱, 짝수면 제곱을 반환하는 함수
f = lambda x: x ** 3 if x % 2 == 1 else x * x
f(2), f(3)
```

```python
f = lambda x, y : x+y
f(3, 4)
```

- Filter 함수 : filter 함수 내 조건을 만족하는 데이터만 filtering

```python
# e.g. 리스트에서 홀수만 골라 새로운 리스트를 만듬
a = [23, 45, 78 ,76, 35, 92]
b = list(filter(lambda x: x % 2 == 1, a))
b
```
